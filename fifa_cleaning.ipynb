```python
# define a function to extract player status
def get_player_status(duration):
    if 'On Loan' in duration:
        return 'On loan'
    elif 'Free' in duration:
        return 'Free'
    else:
        return 'Long term'

# apply the function to create the "player_status" column
raw['Player_status'] = raw['Contract'].apply(get_player_status)
```

```python
 #extract contract start and end dates
# define regular expressions to extract year values
start_regex = r'(\d{4})'
end_regex = r'(On Loan|\d{4})'

# extract year values and assign to new columns
raw['Contract_start'] = raw['Contract'].str.extract(start_regex)
raw['Contract_end'] = raw['Contract'].str.extract(end_regex)

# replace "On Loan" with contract start year
raw['Contract_end'] = raw['Contract_end'].apply(lambda x: raw['Contract_start'] if x == 'On Loan' else x)

# convert columns to datetime
raw['Contract_start'] = pd.to_datetime(raw['Contract_start'], format='%Y')
raw['Contract_end'] = pd.to_datetime(raw['Contract_end'], format='%Y')
#Years only
raw['Contract_end']=raw['Contract_end'].dt.strftime('%Y')
raw['Contract_Start']=raw['Contract_start'].dt.strftime('%Y')
```

```python
# Define a function to convert height to meters
def height_to_meters(height):
    if 'cm' in height:
        cm = int(height.replace('cm', '').strip())
        return round(cm / 100, 2)
    else:
        feet, inches = height.split("'")
        total_inches = int(feet) * 12 + int(inches.strip('"'))
        return round(total_inches * 0.0254, 2)

# Create a new column "height (m)"
raw['Height(m)'] = raw['Height'].apply(height_to_meters)
```

```python
#cleaning Weights column
raw['Weight'].unique()
```

```python
#converting lbs to kgs 
# Define a function to convert lbs to kg
def lbs_to_kg(weight):
    if 'lbs' in weight:
        lbs = int(weight.replace('lbs', '').strip())
        return round(lbs * 0.45359237, 1)
    else:
        return int(weight.replace('kg', '').strip())

# Create a new column "weight (kg)"
raw['Weight(kg)'] = raw['Weight'].apply(lbs_to_kg)
```

```python
# remove Euro sign and extract numeric value and unit
raw['Value'] = raw['Value'].str.replace('€', '')
raw[['value', 'unit']] = raw['Value'].str.extract(r'([\d.]+)\s*([a-zA-Z]+)', expand=True)

# convert value to float and multiply by appropriate factor
raw['value'] = raw['value'].astype(float)
raw.loc[raw['unit'] == 'M', 'value'] *= 1000000
raw.loc[raw['unit'] == 'K', 'value'] *= 1000

# remove 'unit' column and update 'Value' column with new values
raw = raw.drop(columns=['unit'])
raw['value'] = raw['value'].fillna(0)  # replace missing values with zero
raw['value'] = np.where(np.isinf(raw['value']), 0, raw['value'])  # replace infinite values with zero
raw['Value'] = raw['value'].astype(int)  # convert to integer data type
```

```python
# Define the remove_k_and_convert_to_int() function
def remove_k_and_convert_to_int(value):
    if pd.isna(value):
        return pd.NA
    elif value == '0':
        return 0
    elif 'K' in value:
        value = value.replace('K', '')
        value = float(value) * 1000
        value = int(value)
    else:
        try:
            value = int(value)
        except ValueError:
            value = pd.NA
    return value

raw['Hits'] = raw['Hits'].apply(remove_k_and_convert_to_int).astype('Int64')
# Fill NaN values in 'Hits_int' column with 0
raw['Hits'] = raw['Hits'].fillna(0)


```
